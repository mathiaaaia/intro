```python
from google.colab import drive
import yaml
import os
import subprocess
import sys
import importlib.util

# 1. Montage du Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. Chemin vers votre fichier
yml_path = '/content/drive/MyDrive/EPFL/adsml.yml'

def check_if_installed(package_name):
    """Vérifie si un paquet est déjà disponible dans l'environnement."""
    # Ancien comportement: return importlib.util.find_spec(package_name) is not None
    # Nouveau comportement: Vérifie des paquets clés pour une détection plus robuste de l'environnement Colab
    packages_to_check = ['numpy', 'pandas', 'sklearn'] # Ajout de 'sklearn' pour scikit-learn
    return all(importlib.util.find_spec(pkg) is not None for pkg in packages_to_check)

def install_from_yml(path):
    """Optimise l'installation des paquets pour Google Colab en lisant un fichier YAML.
    Cette fonction gère l'exclusion de paquets pré-installés ou problématiques
    et tente une installation groupée avec un fallback individuel en cas d'échec.
    """
    # --- Vérification de base ---
    # On teste 'numpy', 'pandas', 'sklearn' car ils sont essentiels et souvent pré-installés
    if check_if_installed("base_packages_exist"): # Le nom du paquet n'importe plus, seule la fonction check_if_installed est appelée
        print("✅ L'environnement est déjà configuré (numpy, pandas, sklearn détectés). Saut de l'installation.")
        return

    if not os.path.exists(path):
        print(f"❌ Erreur : Le fichier {path} est introuvable.")
        return

    print("🔍 Analyse du fichier adsml.yml...")
    with open(path, 'r') as f:
        data = yaml.safe_load(f)

    dependencies = data.get('dependencies', [])
    pip_packages = []

    # Define EXCLUDED packages
    EXCLUDED = ['python', 'pip', 'nb_black', 'nb_conda', 'sqlite', 'ipython', 'jupyter', 'tensorflow', 'protobuf']

    def clean_and_format(pkg_str):
        pkg_str = pkg_str.strip()
        if '=' in pkg_str and not any(op in pkg_str for op in ['==', '>=', '<=', '>', '<']):
            pkg_str = pkg_str.replace('=', '==')
        if pkg_str.startswith('black'): return 'black'
        # Handle gensim specifically
        if pkg_str.startswith('gensim'): return 'gensim'
        return pkg_str

    for dep in dependencies:
        if isinstance(dep, str):
            pip_packages.append(clean_and_format(dep))
        elif isinstance(dep, dict) and 'pip' in dep:
            for p in dep['pip']:
                pip_packages.append(clean_and_format(p))

    # Filter out excluded packages
    filtered_pip_packages = [pkg for pkg in pip_packages if not any(pkg.startswith(e) for e in EXCLUDED)]

    print(f"⏳ Paquets à installer: {filtered_pip_packages}")
    print(f"⏳ Installation de {len(filtered_pip_packages)} paquets en bloc...")
    try:
        # Installation silencieuse
        subprocess.check_call([sys.executable, "-m", "pip", "install", *filtered_pip_packages, "--quiet"])
        print("✅ Environnement configuré avec succès !")
    except Exception as e:
        print(f"⚠️ Erreur lors de l'installation groupée. Passage en mode individuel pour diagnostiquer...")
        # Fallback si l'installation groupée échoue (ex: conflit de version)
        failed_packages = []
        for pkg in filtered_pip_packages:
             try:
                 subprocess.check_call([sys.executable, "-m", "pip", "install", pkg, "--quiet"])
             except:
                 print(f"❌ Erreur sur {pkg}")
                 failed_packages.append(pkg)
        if failed_packages:
            print(f"⚠️ Terminée avec des erreurs pour : {failed_packages}")

# Lancement
install_from_yml(yml_path)
```

## Comprendre l'enregistrement automatique

### Subtask:

#### Explication:
1.  **Enregistrement automatique sur Google Drive** : Google Colab intègre une fonctionnalité d'enregistrement automatique robuste qui sauvegarde votre travail directement et régulièrement sur Google Drive. Cela signifie que toutes vos modifications (code, sorties, texte) sont stockées dans le cloud.
2.  **Déclenchement de l'enregistrement** : L'enregistrement automatique se produit généralement après chaque modification significative de votre notebook, comme l'exécution d'une cellule, la modification de code, l'ajout ou la suppression de cellules de texte ou de code. Vous pouvez souvent voir un indicateur "Enregistrement..." ou "Enregistré" dans la barre supérieure du notebook.
3.  **Prévention de la perte de données** : Cette fonctionnalité est cruciale pour prévenir la perte de données. En cas de déconnexion de votre session, de fermeture accidentelle de l'onglet du navigateur, ou d'autres interruptions imprévues, la version la plus récente de votre travail est déjà sauvegardée sur Google Drive, vous permettant de reprendre là où vous vous étiez arrêté.
4.  **Recommandation de sauvegarde manuelle** : Bien que l'enregistrement automatique soit très fiable, il est toujours recommandé de comprendre et d'utiliser les options de sauvegarde manuelle. Vous pouvez forcer un enregistrement via `Fichier > Enregistrer` ou `Fichier > Enregistrer une copie dans Drive` pour un contrôle accru, en particulier avant des opérations majeures ou lorsque vous souhaitez créer un point de restauration spécifique.

## Sauvegarde manuelle et options

Google Colab offre plusieurs options pour sauvegarder manuellement votre travail, ce qui est crucial pour gérer différentes versions de votre notebook et pour assurer que vos modifications sont bien enregistrées.

### 1. Fichier > Enregistrer (ou Ctrl+S / Cmd+S)
Cette option permet de **forcer un enregistrement** de l'état actuel de votre notebook. Colab sauvegarde automatiquement votre travail à intervalles réguliers, mais utiliser cette option garantit que toutes les modifications récentes sont immédiatement synchronisées avec la version stockée sur Google Drive. C'est utile avant de fermer le navigateur ou de redémarrer le runtime, ou simplement pour vous assurer que votre progrès est sécurisé.

### 2. Fichier > Enregistrer une copie sur Drive
Cette fonctionnalité est extrêmement utile pour la **gestion des versions** et l'**expérimentation**. Elle crée une **nouvelle copie indépendante** de votre notebook dans Google Drive. Cela vous permet de:
*   **Créer des points de contrôle**: Vous pouvez enregistrer une version stable de votre code avant d'introduire des changements majeurs ou expérimentaux.
*   **Tester de nouvelles idées**: Vous pouvez modifier la copie sans affecter le notebook original.
*   **Partager une version spécifique**: Si vous voulez partager un état précis de votre travail sans que les futures modifications n'affectent la version partagée.
La nouvelle copie est entièrement séparée et ne sera pas automatiquement mise à jour par les modifications apportées au notebook d'origine.

### 3. Enregistrer une copie sur GitHub
Colab s'intègre également avec GitHub pour faciliter le contrôle de version et la collaboration:
*   **Fichier > Enregistrer une copie en tant que GitHub Gist**: Crée un Gist GitHub public (ou privé si vous le spécifiez) de votre notebook. Les Gists sont parfaits pour partager des extraits de code ou des notebooks rapidement sans avoir à gérer un dépôt complet.
*   **Fichier > Enregistrer une copie sur GitHub**: Vous permet de sauvegarder une copie de votre notebook directement dans un dépôt GitHub existant. C'est une excellente pratique pour versionner vos projets et collaborer avec d'autres développeurs.

### 4. Fichier > Historique des révisions
Colab maintient un historique des révisions de votre notebook, vous permettant de revenir à des versions antérieures. En sélectionnant **'Fichier > Historique des révisions'**, vous pouvez:
*   Voir une liste des différentes versions de votre notebook, souvent horodatées et parfois avec les utilisateurs qui ont fait les modifications (en cas de collaboration).
*   Comparer les versions pour identifier les changements.
*   **Restaurer une version antérieure**: Si vous avez fait une erreur ou souhaitez revenir à un état précédent, vous pouvez sélectionner une révision et la restaurer comme version actuelle de votre notebook.

### Où sont sauvegardés les notebooks Colab et comment les retrouver ?

Les notebooks Google Colab sont automatiquement enregistrés par défaut dans un dossier nommé **'Colab Notebooks'**.
Ce dossier se trouve généralement à la racine de votre Google Drive.

Pour retrouver vos notebooks :

1.  **Navigation Manuelle :** Accédez à votre Google Drive (drive.google.com) et recherchez le dossier **'Colab Notebooks'**. Tous vos notebooks Colab y seront listés.
2.  **Fonction de Recherche :** Utilisez la barre de recherche de Google Drive. Vous pouvez taper le nom de votre notebook (même une partie du nom) pour le localiser rapidement. Vous pouvez également rechercher "type:document owner:me Colab Notebooks" pour affiner la recherche.

**Important :** Si vous avez créé un notebook à partir d'un fichier existant stocké ailleurs sur votre Drive, ou si vous l'avez explicitement enregistré dans un dossier spécifique (via 'Fichier > Enregistrer une copie dans Drive' ou 'Fichier > Déplacer'), le notebook se trouvera dans cet emplacement choisi, et non dans le dossier 'Colab Notebooks'.
